##### Test de convergence des économies de la région MENA
###### La convergence vers l'équilibre permanent est l'une des conclusions les plus importantes et les plus optimistes du modèle de Solow : elle implique que les pays pauvres devraient croître plus rapidement et finir par rattraper les pays plus riches, souvent en raison d'un taux de rendement marginal du capital investi plus élevé dans les pays à croissance rapide. Les études empiriques montrent une relation négative entre le taux de croissance moyen et le PIB par habitant donc les pays dont le revenu initial est plus faible connaissent une croissance plus rapidement. Dans cette partie on va essayer d'évaluer de manière critique la pertinence de cette hypothèse de convergence en se basant sur les données de la région MENA. 



#### Hypothèses de test :
H0 : Il existe une relation négative entre le taux de croissance et le PIB par habitant.                                          
H1 : Il n’existe pas de corrélation entre le niveau de richesse d’un pays et son taux de croissance.

```python
# Importation des packages
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import scipy as sc
```

```python
# importation des données
df = pd.read_csv("D:\GDP-PER-CAPITA.csv", index_col = 0)
df
```

```python
sns.regplot(x=df["GDP per capita (current US$) 2000"], y=df["Avg GDP growth rate\n (2000-2019) en %"], fit_reg="true")
plt.show()
```

```python
#Déterminantion du coefficient de corrélation
x = df['GDP per capita (current US$) 2000']
y = df['Avg GDP growth rate\n (2000-2019) en %']
z = df['Country Code']
```

```python
# Scatter plot with plots labels
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
plt.style.use('seaborn-white')
figure(figsize=(8, 6), dpi=100)
plt.xlabel('GDP per capita 2000')
plt.ylabel('Avg GDP growth rate\n (2000-2019) en %')
plt.scatter(x, y, marker='x')
for i, txt in enumerate(z):
    plt.annotate(txt, (x[i], y[i]+0.1))
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b)
```

```python
#Déterminantion du coefficient de corrélation
np.corrcoef(x,y)
```

# Interprétation du test Statistique

### Afin de vérifier l'hypothése de convergence nous avons procéder par un test statistique  de correlation linéaire entre le niveau de PIB par habitant et le taux de croissance annuelle moyen avec python. Ce qui approuve qu'il  n'y a pas de correlation entre les variables étudiées.
