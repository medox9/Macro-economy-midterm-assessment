```python
%matplotlib inline
import pandas as pd
df = pd.read_csv("D:\GDP-PER-CAPITA.csv", index_col = 0)
df
```

```python
x = df['2000']
y = df['Avg GDP growth rate\n (2000-2019) en %']
z = df['Country Code']
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(8, 6), dpi=80)
plt.xlabel('GDP per capita 2000')
plt.ylabel('Avg GDP growth rate\n (2000-2019) en %')
plt.scatter(x, y)
for i, txt in enumerate(z):
    plt.annotate(txt, (x[i], y[i]+0.1))
```
